```python
con_df = pd.read_csv('Street_Closures_due_to_construction_activities_by_Block.csv')
```

```python
def daterange(start_date, end_date):
    for n in range(int ((end_date - start_date).days) +1):
        yield start_date + datetime.timedelta(n)

check=0
date_construction_num = {}

for index, row in con_df.iterrows():
    #if(check==5):
    #    break

    start_date = pd.Timestamp(row["WORK_START_DATE"])
    end_date = pd.Timestamp(row["WORK_END_DATE"])

    for n in range(int ((end_date - start_date).days) +1):
        #print((start_date+datetime.timedelta(n)).date())
        date = (start_date+datetime.timedelta(n)).date()
        if(date not in date_construction_num):
            date_construction_num[date] = {'M':0, 'X':0, 'B':0, 'Q':0, 'S':0}
        
        date_construction_num[date][row["BOROUGH_CODE"]] =  date_construction_num[date][row["BOROUGH_CODE"]] + 1
    
    check = check + 1
```

```python
#M: manhattan, X: Bronx, B: Brooklyn, Q: Queens, S: Staten Island
M_arr = [0]*df.shape[0]
X_arr = [0]*df.shape[0]
B_arr = [0]*df.shape[0]
Q_arr = [0]*df.shape[0]
S_arr = [0]*df.shape[0]
```

```python
df.head
```

```python
for index, row in df.iterrows():
    date = pd.Timestamp(row["tpep_pickup_datetime"]).date()
    if(date not in date_construction_num):
        print(date)
        continue
    M_arr[index] = date_construction_num[date]["M"]
    X_arr[index] = date_construction_num[date]["X"]
    B_arr[index] = date_construction_num[date]["B"]
    Q_arr[index] = date_construction_num[date]["Q"]
    S_arr[index] = date_construction_num[date]["S"] 
```

```python
df['M'] = pd.Series(M_arr, index=df.index)
```

```python
df['X'] = pd.Series(X_arr, index=df.index)
df['B'] = pd.Series(B_arr, index=df.index)
df['Q'] = pd.Series(Q_arr, index=df.index)
df['S'] = pd.Series(S_arr, index=df.index)
```
